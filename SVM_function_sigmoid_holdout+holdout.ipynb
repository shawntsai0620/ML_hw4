```python
#Basic packages always been used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Data preprocessing useful function
from sklearn.preprocessing import LabelEncoder
# Function for spilting training & testing data set
from sklearn.model_selection import train_test_split
# Algorithm
from sklearn import svm
from sklearn.model_selection import KFold
# Functions for evaluation
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import plot_roc_curve, roc_auc_score, roc_curve
```

```python
df = pd.read_csv('C:/Users/NelsonTsai/Desktop/ML_hw3/DATA_test.csv')
df
```

```python
x=df.drop(['MonkeyPox_Negative','MonkeyPox_Positive'],axis=1).copy()
x.head()
```

```python
y=df['MonkeyPox_Positive'].copy()
y.head()
```

```python
#kernel = rbf
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
# 5.2 Manually peforming hyper parameter tuning
svm_c = [1]#hyper parameter:C
svm_gamma = [0.01,0.05,0.1]#hyper parameter:gamma
seed = [100,400,800]
sum_1 = 0
sum_2 = 0
sum_3 = 0
for i in svm_c:
    for j in svm_gamma:
        for l in seed:
            x_train,x_validation,y_train,y_validation=train_test_split(x_train,y_train,test_size=0.2,random_state=l)
            rbfModel=svm.SVC(kernel='sigmoid',C=i,gamma=j, random_state=4)
            rbfModel.fit(x_validation,y_validation)
            validation_prediction=rbfModel.predict(x_validation)
            validation_acc=accuracy_score(y_validation,validation_prediction)
            if i == 1 and j == 0.01 :
                sum_1 += validation_acc
            if i == 1 and j == 0.05 :
                sum_2 += validation_acc
            if i == 1 and j == 0.1 :
                sum_3 += validation_acc
print('validation result:')
print('validation ACC:',round((np.mean(sum_1))*100,2)/3)
print('validation ACC:',round((np.mean(sum_2))*100,2)/3)
print('validation ACC:',round((np.mean(sum_3))*100,2)/3)
print('')
```
